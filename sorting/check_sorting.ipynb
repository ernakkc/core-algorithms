# Check if an Array is Sorted

```python
def isSorted(array):
    n = len(array)

    for i in range(n-1):
        if array[i+1] != None and array[i] <= array[i+1]:
            continue
        else:
            return False
    return True
```

### Time Complexity: 
- O(n) where n is the number of elements in the array.

### Examples

```python
array = [10, 20, 30, 40, 50]
isSorted(array)
```

```python
array = [2,5,7,1,5]
isSorted(array)
```
