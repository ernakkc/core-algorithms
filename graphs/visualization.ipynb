```python
import networkx as nx
import matplotlib.pyplot as plt

# Creating a Undirected Graph
G = nx.Graph()

# Adding enges
edges = [('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'E'), ('C', 'F'), ('C', 'G')]
G.add_edges_from(edges)

layout = nx.spring_layout(G)

# Grafı çizdiriyoruz
nx.draw(
    G, 
    pos=layout, 
    with_labels=True,           # Write name of nodes (A, B, vs.) 
    node_color='lightblue',     # Node color
    node_size=1500,             # Node size
    edge_color='gray',          # Edge color
    font_size=14,               # Font Size
    font_weight='bold'          # Font type
)

# Show the graph
plt.title("Graph Visualization")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edge('A', 'B', weight=4)
G.add_edge('A', 'C', weight=2)
G.add_edge('B', 'D', weight=5)
G.add_edge('C', 'D', weight=1)
G.add_edge('C', 'E', weight=6)
G.add_edge('D', 'E', weight=3)


shortest_path = nx.dijkstra_path(G, source='A', target='E', weight='weight')
print(f"Shortest path from A to E: {shortest_path}")

pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_color='lightgray', node_size=1500)

way_edges = list(zip(shortest_path, shortest_path[1:]))

nx.draw_networkx_nodes(G, pos, nodelist=shortest_path, node_color='lightgreen', node_size=1500)

nx.draw_networkx_edges(G, pos, edgelist=way_edges, edge_color='green', width=3)

edge_wheights = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_wheights)

plt.title("Dijkstra Algorithm Visualization")
plt.show()
```
